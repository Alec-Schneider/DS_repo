KFold Cross Validation:
- More thorough way of cross validation
- split training set into K different groups
- for k folds, train your data on K-1 groups, and score it on the one remaining set
- is only used on the training set

Ex:
If there are 5 fold CV, you split the datset into 5 different stratifications. For that # of folds, you have that many rounds of CV. Every fold serves as training and validation data, but not at the same time

Look ahead KFold CV makes sure to keep scoring based on what's coming next
