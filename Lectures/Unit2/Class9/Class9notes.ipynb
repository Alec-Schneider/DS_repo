```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../data/titanic.csv')
```

```python
sns.catplot(x='Pclass', y='Age', data=data)
```

```python
sns.catplot(x='Pclass', y='Age', data=data, kind='swarm')
```

```python
sns.catplot(x='Pclass', y='Age', data=data, kind='violin')
```

```python
sns.catplot(x='Pclass', y='Age', data=data, kind='boxen')
```

```python
sns.catplot(x='Pclass', y='Age', kind='swarm', hue='Sex', data=data)
```

```python
sns.catplot(x='Pclass', y='Age', kind='swarm', col='Sex', data=data)
```

```python
sns.lmplot(x='Pclass', y='Survived', data=data)
```

```python
sns.distplot(data['Age'].fillna(data['Age'].mean()))
```

```python
import matplotlib.pyplot as plt
sns.jointplot(x='Age', y='Fare', data=data[data.Sex=='male'])
# plt.title('Age vs Fare for Males')
```

```python
ax = sns.distplot(data[data.Sex == 'female']['Age'].fillna(0), label='females')
sns.distplot(data[data.Sex == 'male']['Age'].fillna(0), label='males', ax=ax)
plt.legend()
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data.corr())
```
